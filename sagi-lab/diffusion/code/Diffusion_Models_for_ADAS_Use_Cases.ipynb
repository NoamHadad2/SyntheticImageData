```python


import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

prompt = "A photograph taken from standing stright position on an asphalt road, looking at a fallen tree log blocking the entire highway ahead."
prompt = "A realistic photograph from a pedestrian viewpoint on a residential suburban street. A large fallen tree log is completely blocking the asphalt road. Houses, parked cars, and sidewalks line the street on both sides."
print(f"Generating image for: {prompt}")
image = pipe(prompt).images[0]

display(image)
image.save("sagi_demo_image.png")
```
